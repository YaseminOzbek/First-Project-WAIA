```python
# Replace the ' 's with NaN
df = df.replace(" ",np.NaN)
```

```python
# Impute the missing values with mean imputation
df = df.fillna(df.mean())
```

```python
# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

```python
# To check missing values
df.isnull().sum()
```

```python
df['gender'].value_counts()
```

```python
#Counting all vales here
df['gender'].value_counts(dropna=False)
```

```python
#Females have more friends than males
plt.figure(figsize=(15,8))
sns.barplot(x=df['age'],y=df['friend_count'],hue=df.gender)
```

```python
#divided the age into a group of 10. see last column
labels=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
df['age_group'] = pd.cut(df.age,bins=np.arange(10,121,10),labels=labels,right=True)
df.head()
```

```python
#Counting value in age groups
df.age_group.value_counts()
```

```python
#Females have more friends than males 
#With age-group we have had a more clear map  
plt.figure(figsize=(15,8))
sns.barplot(x=df['age_group'],y=df['friend_count'],hue=df.gender)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['age_group'],y=df['friendships_initiated'],hue=df.gender)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['age_group'],y=df['tenure'],hue=df.gender)
```

```python
#Females spent more time than males.
sns.barplot(x=df['gender'],y=df['tenure'],hue=df.gender, hue_order = ['male', 'female'])
plt.title('Gender and Tenure')
```
